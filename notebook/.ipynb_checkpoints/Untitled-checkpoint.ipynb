```python
import os
import requests
from flask import Flask, render_template, jsonify
```

```python
def get_files_recursive(access_token,owner,repo,path='',allowed_extensions=None):
    if allowed_extensions is None:
        allowed_extensions = []
        
    github_enterprise_base_url = 'https://git.cglcloud.com/api/v3'
    github_api_url = f'{github_enterprise_base_url}/repos/{owner}/{repo}/contents/{path}'
    
    headers = {'Authorization':"Token "+access_token, 'Accept':'application/vnd.github+json'}
    response = requests.get(github_api_url,headers=headers)
    contents= response.json()
    

    files = []
    print(contents)
    for content in contents:
        if content['type'] == 'dir':
            sub_path = os.path.join(path,content['name']).replace("\\","/")
            files.extend(get_files_recursive(access_token,owner,repo,sub_path,allowed_extensions))
        else:
            file_extension = os.path.splitext(content['name'])[1]
            if file_extension in allowed_extensions:
                file_data = {'name': content['name'], 'url': content['download_url'],'extension': file_extension}
                files.append(file_data)

    return files
```
